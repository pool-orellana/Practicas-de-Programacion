```python
print('Define el equipamiento para una aventura de trekking y su valor en trekjuls (moneda del juego):')
```

```python
#Equipamiento - Costo en monedas trekjuls

gps = 30
botiquin = 30
enlatados = 13.5
agua = 10
vestimenta = 60.8
navaja_multiuso = 17.8
botas = 30.5
linterna = 8.5
saco_de_dormir = 50
```

```python
# Lista los elementos

print('Lista tres objetos del equipamiento: Nombre y valor')
print('gps = ', gps)
print('botas = ', botas)
print('saco_de_dormir =', saco_de_dormir)
```

```python
# Combinacion de elementos

print("¿Puedes convinar elementos de tu equipo para prepararte mejor?")
alimentacion = enlatados + agua
equipo_seguridad = navaja_multiuso + botiquin

print('Alimentacion: ', alimentacion)
print('Equipo de seguridad', equipo_seguridad)
```

```python
# Comparacion de elementos

print("¿El precio de la alimentacion es menor que del equipo de seguridad?")
print(alimentacion < equipo_seguridad)
```

```python
# Operacion aritmetica de elementos

print('Cuanto valdira comprar la vestimenta + el saco de dormir')
print('El costo final de los dos elementos serian: ', vestimenta + saco_de_dormir,' trekjuls')
```

```python
# Comparacion de elementos

print('Si tengo 10 trekjuls, ¿Me alcanzara para compar mas agua?')
print(agua >= 10)
```

**Clase 2: Tomando decisiones.** (Estructuras de control)

```python
print('La esfinge te dice: No debes colocar más de 5 objetos en tu mochila,' 
'y tampoco pasarte de 130 trekjuls: ¿Cuales elementos colocarías?')

mochila = 0

if alimentacion + equipo_seguridad + botas <= 130:
  mochila = alimentacion + equipo_seguridad + botas
  print("El valor de la mochila es inferior a los 200 trekjuls", 'Valor: ', mochila)

elif alimentacion + equipo_seguridad + vestimenta <= 130:
  mochila = alimentacion + equipo_seguridad + vestimenta
  print("El valor de la mochila es inferior a los 200 trekjuls", 'Valor: ', mochila)

else:
  print('Se necesitan mas trekjuls para adquirir elementos.')
```

```python
print ("Es tu dia de suerte! Te voy a dar otra mochila, pero solo puedes agregarle alimentacion")

mochila_apoyo = 0

while mochila_apoyo <= 260:
  mochila_apoyo += alimentacion

print('Mochila de apoyo: ', mochila_apoyo)

mochila_apoyo -= alimentacion
print('Necesitamos quitar un enlatado')
```
