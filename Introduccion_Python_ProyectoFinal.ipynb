**Clase 2: Tomando decisiones.** (Estructuras de control)

*Es hora de comenzar a equiparte, por eso la esfinge te pedirá
adicionar elementos a tu mochila que no superen cierto monto
en su costo, es un buen momento para aplicar los condicionales.
Como la esfinge ve que podría hacerte falta un elemento, te regala
una nueva mochila, pero solo para agregarle agua en botella, y
claro ella no te va regalar nada sin antes ponerte un limite, por
eso el uso de ciclos te ayudará a validar y ahorrar esfuerzo para
cumplir las restricciones que te pongan.*

```python
print('La esfinge te dice: No debes colocar más de 5 objetos en tu mochila,' 
'y tampoco pasarte de 130 trekjuls: ¿Cuales elementos colocarías?')

mochila = 0

if alimentacion + equipo_seguridad + botas <= 130:
  mochila = alimentacion + equipo_seguridad + botas
  print("El valor de la mochila es inferior a los 200 trekjuls", 'Valor: ', mochila)

elif alimentacion + equipo_seguridad + vestimenta <= 130:
  mochila = alimentacion + equipo_seguridad + vestimenta
  print("El valor de la mochila es inferior a los 200 trekjuls", 'Valor: ', mochila)

else:
  print('Se necesitan mas trekjuls para adquirir elementos.')
```

```python
print ("Es tu dia de suerte! Te voy a dar otra mochila, pero solo puedes agregarle alimentacion")

mochila_apoyo = 0

while mochila_apoyo <= 260:
  mochila_apoyo += alimentacion

print('Mochila de apoyo: ', mochila_apoyo)

mochila_apoyo -= alimentacion
print('Necesitamos quitar un enlatado')
```

```python
print("Le hice una actualización a tu mochila te dice la esfinge, porque solo podias conocer el valor de los objetos que tenias")
print("Ahora sabras el objeto que tienes, la cantidad y su valor unitario, pero tu debes volverla a llenarla")

mochila_actualizada = {
    "alimentacion" : {"cantidad": 1, "valor_unitario": alimentacion},
    "equipo_seguridad" : {"cantidad": 1, "valor_unitario": equipo_seguridad},
    "gps" : {"cantidad": 1, "valor_unitario": gps}
}

mochila_dos_actualizada = {
    "alimentacion" : {"cantidad": 2, "valor_unitario": alimentacion}
}
```

**Clase 4: Organizandonos un poco.** (Funciones)


*Es momento de organizar un poco nuestro juego ‘trekjuls’, y
qué mejor forma de hacerlo que con el uso de funciones. En el
último reto la esfinge te pedirá que agregues más elementos a las
mochilas de tu equipo, pero solo te dejará hacerlo si logras calcular
el total de elementos que hayan por todas las mochilas, con eso
ya estarás listo para embarcar la aventura del trekking y habrás
superado todos los retos*

```python
print('La esfinge te pide que para cuatro integrantes se recolecten 3 elementos sin importar las cantidades '
'y te da las siguientes opciones: brujula, linterna_funcional, snacks y agua_en_botella')
print('Ademas te va a pedir calcular el total de elementos que hay en tu equipo')


def agregarElementosAMochilas():
    for compartimiento in range(3):
        vehiculo[compartimiento] = {
            "alimentacion" : {"cantidad": 1, "valor_unitario": alimentacion},
            "equipo_seguridad" : {"cantidad": 2, "valor_unitario": equipo_seguridad},
            "gps" : {"cantidad": 3, "valor_unitario": gps},
            "navaja_multiuso" : {"cantidad": 2, "valor_unitario": navaja_multiuso},
        }
    imprimir(vehiculo)
    calcularTotalElementosEnMochilas(vehiculo)

def calcularTotalElementosEnMochilas(vehiculo):
    total_elementos_mochilas = {}
    
    print("Calculo elementos en mochila")
    for mochila in vehiculo: 
        for elemento, detalle in mochila.items(): 
            if elemento in total_elementos_mochilas:
                total_elementos_mochilas[elemento] += detalle["cantidad"]
            else:
                total_elementos_mochilas[elemento] = detalle["cantidad"]

    print(total_elementos_mochilas)


def imprimir(estructura):
    for objeto in estructura:
        print(objeto)

agregarElementosAMochilas()
```
